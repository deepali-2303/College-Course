```python
data = {'Length (cm)': [], 'Label (0->unripe 1->ripe)': []}
with open('data_binary.csv', newline='') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        data['Length (cm)'].append(float(row['Length (cm)']))
        data['Label (0->unripe 1->ripe)'].append(int(row['Label (0->unripe 1->ripe)']))

lengths = data['Length (cm)']
labels = data['Label (0->unripe 1->ripe)']

mean_length = sum(lengths) / len(lengths)
mean_label = sum(labels) / len(labels)

numerator = sum([(x - mean_length) * (y - mean_label) for x, y in zip(lengths, labels)])
denominator = sum([(x - mean_length) ** 2 for x in lengths])
slope = numerator / denominator

intercept = mean_label - slope * mean_length

print("Slope (m):", slope)34
print("Intercept (b):", intercept)
```

```python
def logistic_regression_predictor(x,slope,intercept):
    y = intercept+x*slope
    P = 1/(1+np.exp(1)**(-y))
    if P>0.5:
        print("The banana of length",x,"cm is ripe")
    else:
        print("The banana of length",x,"cm is not ripe")
    #print("The probabilty of a banana of length",x,"cm being ripe is",P,"hence, it is Ripe")
```

```python
#input values
input = [19.8, 16.5, 15.7, 12.82, 14.23, 17, 14.9, 10.3]
for i in input:
    logistic_regression_predictor(i,slope,intercept)
```

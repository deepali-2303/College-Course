```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('/Users/arshraina/Desktop/Desktop/College/Sem6/ML/Experiment 2/data.csv', sep=',')
```

```python
df
```

```python
# Scatter Plot
plt.figure(figsize=(10, 6))
scatter_plot = sns.scatterplot(x='Training Starting Age', y='Peak Rating', data=df, s=100)

plt.title('Scatter Plot of Training Starting Age vs. Peak Rating')
plt.xlabel('Training Starting Age')
plt.ylabel('Peak Rating')

plt.show()
```

```python
# Linear Regression
training_starting_age = df['Training Starting Age']
peak_rating = df['Peak Rating']

mean_x = sum(training_starting_age) / len(training_starting_age)
mean_y = sum(peak_rating) / len(peak_rating)

SOP_deviations_xy = sum((x - mean_x) * (y - mean_y) for x, y in zip(training_starting_age, peak_rating))
Square_deviation_x = sum((x - mean_x) ** 2 for x in training_starting_age)

slope = SOP_deviations_xy/Square_deviation_x
intercept = mean_y - slope * mean_x
print("slope: ",slope)
print("intercept: ",intercept)

predictions = [slope * x + intercept for x in training_starting_age]

plt.figure(figsize=(10, 6))

plt.scatter(training_starting_age, peak_rating, s=50, label='Original Data Points')

plt.plot(training_starting_age, predictions, color='red', linewidth=3, label='Linear Regression Line')

plt.title('Linear Regression: Training Starting Age vs. Peak Rating')
plt.xlabel('Training Starting Age')
plt.ylabel('Peak Rating')
plt.legend()

plt.show()
```

```python
# Least Squared Error
error = []
for i in range(len(training_starting_age)):
    e = peak_rating[i]-predictions[i]
    error.append(e)

df_new = pd.DataFrame({
    'Training Starting Age': training_starting_age,
    'Peak Rating': peak_rating,
    'Predictions': predictions,
    'Error': error
})
```

```python

plt.figure(figsize=(10, 6))

plt.scatter(training_starting_age, peak_rating, s=50, label='Original Data Points')

plt.scatter(training_starting_age, predictions, s=50, label='Predicted Data Points', color='black')
for x, y_actual, y_pred in zip(training_starting_age, peak_rating, predictions):
    plt.plot([x, x], [y_actual, y_pred], linestyle='dotted', color='gray')

plt.plot(training_starting_age, predictions, color='red', linewidth=3, label='Linear Regression Line')


plt.title('Least Squared Regression: Training Starting Age vs. Peak Rating')
plt.xlabel('Training Starting Age')
plt.ylabel('Peak Rating')
plt.legend()

plt.show()
```

```python
x_input = 16
y_predicted = slope*x_input+intercept
print("Peak Rating Prediction:",y_predicted)
```

```python
#Testing with Scikit-learn
training_starting_age = df['Training Starting Age'].values.reshape(-1, 1)
peak_rating = df['Peak Rating'].values

model = LinearRegression()

model.fit(training_starting_age, peak_rating)

slope = model.coef_[0]
intercept = model.intercept_

print("slope: ", slope)
print("intercept: ", intercept)

predictions = model.predict(training_starting_age)

plt.figure(figsize=(10, 6))
plt.scatter(training_starting_age, peak_rating, s=50, label='Original Data Points')
plt.plot(training_starting_age, predictions, color='red', linewidth=3, label='Linear Regression Line')
plt.title('Scikit-learn implementation: Training Starting Age vs. Peak Rating')
plt.xlabel('Training Starting Age')
plt.ylabel('Peak Rating')
plt.legend()
plt.show()
```

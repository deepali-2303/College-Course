```python
df=pd.read_csv("data.csv")
df
```

```python
sns.scatterplot(x="Height",
                    y="Weight",
                    data=df)
```

```python
sns.regplot(x='Height', y='Weight', data=df)
```

```python
import numpy as np
import matplotlib.pyplot as plt


height = df['Height']
weight = df['Weight']


learning_rate = 0.01
epochs = 10000


slope = 0.5  
intercept = 50  

errors = []


for epoch in range(epochs):
   
    predicted_weight = slope * height + intercept

    
    d_slope = (-2/len(height)) * np.sum(height * (weight - predicted_weight))
    d_intercept = (-2/len(height)) * np.sum(weight - predicted_weight)

    error = np.sum((predicted_weight - weight) ** 2)
    errors.append(error)

    
    slope -= learning_rate * d_slope
    intercept -= learning_rate * d_intercept


test_heights = np.array([1.70, 1.75, 1.85])
predicted_weights = slope * test_heights + intercept


plt.plot(range(epochs), errors, label='Least Squares Error')
plt.xlabel('Epochs')
plt.ylabel('Error')
plt.title('Least Squares Error over Epochs')
plt.legend()
plt.show()


plt.scatter(height, weight, label='Actual data')


plt.plot(height, slope * height + intercept, color='red', label='Regression line')


plt.scatter(test_heights, predicted_weights, color='green', label='Predicted data')


plt.scatter(height, weight, label='Actual data', color='blue')
plt.plot(height, predicted_weight, color='red', linestyle='dashed', label='Predicted values')

# Connecting lines to show errors
for i in range(len(height)):
    plt.plot([height[i], height[i]], [weight[i], predicted_weight[i]], color='black', linestyle='dashed')


plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')
plt.legend()

  
plt.show()
```
